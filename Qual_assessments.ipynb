```bash
quast.py -o ./\
-r ../../GCF_003324715.1_ASM332471v1_genomic.fna \
-g ../../GCF_003324715.1_ASM332471v1_genomic.gff \
-t 20\
--k-mer-stats\
--circos \
--pe1 ../../Autocycler_assembly/reads_qc/illumina_1.fastq.gz \
--pe2 ../../Autocycler_assembly/reads_qc/illumina_2.fastq.gz \
--nanopore ../../Autocycler_assembly/ont_6k.fastq \
../../DSM158_NCBI/DSM158.fasta
```

## Merqury assessment for DSM158

```bash
meryl k=21 count illumina_1.fastq.gz output reads_1.meryl
meryl k=21 count illumina_2.fastq.gz output reads_2.meryl
meryl union-sum reads_1.meryl reads_2.meryl output reads.meryl
merqury.sh reads.meryl/ ../../../Autocycler_assembly/final/Cereibacter_sphaeroides_reoriented.fasta DSM158
```
